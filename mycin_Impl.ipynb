```python
import numpy as n
import matplotlib.pyplot as p
```

```python
x = n.arange(-1,1,0.1)
y=n.arange(-1,1,0.1)
def mycin_func(x,y):
    if x > 0 and y>0:
        return x+y-x*y
    elif x<0 and y<0:
        return x+y+x*y
    elif abs(x)==1 and abs(y)==1:
        return 1
    else:
        return (x+y)/(1-min(abs(x),abs(y)))
```

```python
diag = p.figure()
t= p.axes(projection='3d')
for i in x:
    for k in y:
        output_z = mycin_func(i,k)
        t.scatter3D(i,k,output_z)
p.show()
```
